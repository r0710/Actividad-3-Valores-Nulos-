```python
#Primero importamos las bibliotecas que ocuparemos 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Ahora hacemos que el programa lea el csv para que podamos empezar a manipularlo
df_original = pd.read_csv('Boston_United States.csv')
df_original
```

```python
#Borramos esa colmna que nos es innecesaria
df_original.drop(['Unnamed: 0'], axis=1, inplace=True)
df_original
```

```python
#Comprobamos el tipo de datos que tenemos de acuerdo a cada columna
df_original.info()
```

```python
#En este paso me acorde de los arreglos que le hicimos a las columnas de precio y tasa de acepetación, y al ser de gran peso para el estudio de la base. Entonces decicidí hacerselos aquí igual
df_original['host_acceptance_rate']=df_original['host_acceptance_rate'].str.replace('%', '')
df_original['host_acceptance_rate']= df_original['host_acceptance_rate'].astype(float)
df_original['price']=df_original['price'].str.replace('$', '')
df_original['price']=df_original['price'].str.replace(',', '')
df_original['price']= df_original['price'].astype(float)
```

```python
#Creamos 5 copias para cada uno de los metódos que aplicaremos para reemplazar los valores nulos
df_cop1 = df_original.copy()
df_cop2 = df_original.copy()
df_cop3 = df_original.copy()
df_cop4 = df_original.copy()
df_cop5 = df_original.copy()
```

```python
#Definimos una función que nos ayude a contabilizar los valores nulos por columna y en total del dataframe que le entreguemos
def val_nu(dataframe):
    valo_nu_col = dataframe.isnull().sum()
    valo_nu_tot = dataframe.isnull().sum().sum()

    return('Valores nulos por columna:',valo_nu_col,
           "Valores nulos en total:", valo_nu_tot)

val_nu(df_original)
```

```python
#Definimos la función para la sustictución de valores nulos por la mediana (variables cualitativas)
def rempl_vn_mediana(dataframe):
    #Vamos a separar columnas cuantitativas del data frame
    col_cuan = dataframe.select_dtypes(include=['float64', 'int64', 'float', 'int'])
    #Ahora separamos las columnas con datos cualitativas
    col_cual = dataframe.select_dtypes(include=['object', 'datetime', 'category'])
    #Sustituimos las variables cuantitativas por promedios
    cuan = col_cuan.fillna(round(col_cuan.median(), 1))
    #Unimos el data frame
    dt_f = pd.concat([cuan, col_cual], axis=1)
    return(dt_f)

#Ahora solo almacenamos en una variable el nuevo dataframe que creará el programa, y para esto le entregamoa una de las copias que creamos anteriormente
pun1 = rempl_vn_mediana(df_cop1)
#Comprobamos que las variables de tipo cuantitativas no contengan valores nulos
val_nu(pun1)
```

```python
#Para tener mayor certez recatificamos la información que contiene cada columna
pun1.info()
```

```python
#Pasamos el data frame a un csv
pun1.to_csv('Primer_punto.csv')
```

```python
#Definimos la función para la sustictución de valores nulos con el método ffill (variables cuantitativas)
def rempl_vn_ffill(dataframe):
    #Vamos a separar columnas cuantitativas del data frame
    col_cuan = dataframe.select_dtypes(include=['float64', 'int64', 'float', 'int'])
    #Ahora separamos las columnas con datos cualitativas
    col_cual = dataframe.select_dtypes(include=['object', 'datetime', 'category'])
    #Sustituimos las variables mediante el método ffill
    cual = col_cual.fillna(method='ffill')
    #Unimos el data frame
    dt_f = pd.concat([col_cuan,cual], axis=1)
    return(dt_f)

#Ahora solo almacenamos en una variable el nuevo dataframe que creará el programa, y para esto le entregamoa una de las copias que creamos anteriormente
pun2 = rempl_vn_ffill(df_cop2)
#Comprobamos que las variables de tipo cuantitativas no contengan valores nulos
val_nu(pun2)
```

```python
#Para tener mayor certez recatificamos la información que contiene cada columna
pun2.info()
```

```python
#Pasamos el data frame a un csv
pun2.to_csv('Segundo_punto.csv')
```

```python
#Definimos la función para la sustictución de valores nulos con el método bfill (variables cuantitativas)
def rempl_vn_bfill(dataframe):
    #Vamos a separar columnas cuantitativas del data frame
    col_cuan = dataframe.select_dtypes(include=['float64', 'int64', 'float', 'int'])
    #Ahora separamos las columnas con datos cualitativas
    col_cual = dataframe.select_dtypes(include=['object', 'datetime', 'category'])
    #Sustituimos las variables mediante el método bfill
    cual = col_cual.fillna(method='bfill')
    #Unimos el data frame
    dt_f = pd.concat([col_cuan,cual], axis=1)
    return(dt_f)
#Ahora solo almacenamos en una variable el nuevo dataframe que creará el programa, y para esto le entregamoa una de las copias que creamos anteriormente
pun3 = rempl_vn_bfill(df_cop3)
#Comprobamos que las variables de tipo cuantitativas no contengan valores nulos
val_nu(pun3)
```

```python
#Pasamos el data frame a un csv
pun3.to_csv('Tercer_punto.csv')
```

```python
#Definimos la función para la sustictución de valores nulos con un str (variables cualitativas)
def rempl_vn_str(dataframe):
    #Vamos a separar columnas cuantitativas del data frame
    col_cuan = dataframe.select_dtypes(include=['float64', 'int64', 'float', 'int'])
    #Ahora separamos las columnas con datos cualitativas
    col_cual = dataframe.select_dtypes(include=['object', 'datetime', 'category'])
    #Sustituimos las variables cuantitativas por un str, y ya que como tenemos diferentes tipos de str en cada columna decicdí hacerlo por un espacio en blanco ya que despues de comprobar los datos nulos que reconocía el programa el espacio no lo detecta como nulo
    cual = col_cual.fillna(" ")
    #Unimos el data frame
    dt_f = pd.concat([col_cuan,cual], axis=1)
    return(dt_f)

#Ahora solo almacenamos en una variable el nuevo dataframe que creará el programa, y para esto le entregamoa una de las copias que creamos anteriormente
pun4 = rempl_vn_str(df_cop4)
#Comprobamos que las variables de tipo cuantitativas no contengan valores nulos
val_nu(pun4)
```

```python
#Pasamos el data frame a un csv
pun4.to_csv('Cuarto_punto.csv')
```
